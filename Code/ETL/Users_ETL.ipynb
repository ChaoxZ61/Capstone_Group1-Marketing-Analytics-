```
user_data = spark.read.format('json').options(lines=True).load('/mnt/group1/usersin/yelp_academic_dataset_user.json')
user_data.show()
```

```
from pyspark.sql.functions import col

# Drop Friends column. Due to nature of dataset, we would not gain much info from it as not all the friends are likely to be in the dataset
user_data = user_data.drop(col('friends'))
user_data = user_data.drop(col('elite'))
user_data.first()
```

```
# Drop null values and duplicate rows
user_data = user_data.dropna().dropDuplicates()
```

```
#Rename Columns 
user_data = user_data.withColumnRenamed("average_stars", "AvgStars")\
    .withColumnRenamed("compliment_cool", "CompCool")\
    .withColumnRenamed("compliment_cute", "CompCute")\
    .withColumnRenamed("compliment_funny", "CompFunny")\
    .withColumnRenamed("compliment_hot", "CompHot")\
    .withColumnRenamed("compliment_list", "CompList")\
    .withColumnRenamed("compliment_more", "CompMore")\
    .withColumnRenamed("compliment_note", "CompNote")\
    .withColumnRenamed("compliment_photos", "CompPhotos")\
    .withColumnRenamed("compliment_plain", "CompPlain")\
    .withColumnRenamed("compliment_profile", "CompProfile")\
    .withColumnRenamed("compliment_writer", "CompWriter")\
    .withColumnRenamed("cool", "Cool")\
    .withColumnRenamed("fans", "Fans")\
    .withColumnRenamed("funny", "Funny")\
    .withColumnRenamed("name", "Name")\
    .withColumnRenamed("review_count", "ReviewCount")\
    .withColumnRenamed("useful", "Useful")\
    .withColumnRenamed("user_id", "UserID")\
    .withColumnRenamed("yelping_since", "YelpingSince")
```

```
# Write Dataframe to JSON in Data Lake
user_data.write.format('json').mode('overwrite').save(
    '/mnt/group1/educationin/user_cleaned.json')
```
